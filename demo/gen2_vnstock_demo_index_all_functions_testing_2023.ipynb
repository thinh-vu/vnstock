```python
df = listing_companies() # G√°n b·∫•t c·ª© h√†m n√†o cho 1 bi·∫øn ƒë·ªÉ xu·∫•t d·ªØ li·ªáu, v√≠ d·ª• ·ªü ƒë√¢y l√† listing_companies
sheet_file = 'listing_companies' # ƒê·∫∑t t√™n cho file Google Sheets s·∫Ω t·∫°o (n·∫øu t·∫°o m·ªõi, d√πng file c√≥ s·∫µn th√¨ kh√¥ng c·∫ßn)
```

```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
from gspread_dataframe import set_with_dataframe

creds, _ = default()
gc = gspread.authorize(creds)
```

### Update an entire worksheet with a DataFrame.

```python
sh = gc.create(sheet_file)
worksheet = gc.open(sheet_file).sheet1 # M·ªü sheet m·∫∑c ƒë·ªãnh ƒë·ªÉ xu·∫•t d·ªØ li·ªáu, ch·ªçn 1 trong 3 t√πy ch·ªçn v·ªõi bi·∫øn worksheet, d√πng d√≤ng n√†o b·ªè comment d√≤ng ƒë√≥, v√† comment d√≤ng kh√¥ng c·∫ßn d√πng
# worksheet = sh.add_worksheet(title="listing_companies", rows="1000", cols="20") # T·∫°o sheet m·ªõi ƒë·ªÉ xu·∫•t d·ªØ li·ªáu
# worksheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/1vT6o6U1dHYMdHASZNKPvmzDRhgKTvQXsZ_46CvBOpuI').sheet1  # M·ªü file sheet c√≥ s·∫µn b·∫±ng URL
set_with_dataframe(worksheet, df)
```

### Update a cell range with a DataFrame

```python
worksheet = gc.open(sheet_file).sheet1 # M·ªü sheet m·∫∑c ƒë·ªãnh ƒë·ªÉ xu·∫•t d·ªØ li·ªáu, ch·ªçn 1 trong 3 t√πy ch·ªçn v·ªõi bi·∫øn worksheet, d√πng d√≤ng n√†o b·ªè comment d√≤ng ƒë√≥, v√† comment d√≤ng kh√¥ng c·∫ßn d√πng

cell_list = worksheet.range('A1:C2') # Ch·ªçn v√πng d·ªØ li·ªáu tr√™n Google Sheets b·∫°n mu·ªën ch√®n d·ªØ li·ªáu t·ª´ DataFrame
# Update the data in the worksheet

for cell, value in zip(cell_list, df.values.flatten()):
    cell.value = value

worksheet.update_cells(cell_list)
```
