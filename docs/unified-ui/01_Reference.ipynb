### Khởi động

````python
from vnstock import Reference
from IPython.display import display

# Initialize Reference layer
ref = Reference()
````

### 1. Company Domain (Thông Tin Công Ty)

Cung cấp hồ sơ doanh nghiệp, ban lãnh đạo, cổ đông lớn, công ty con, sự kiện và biên lợi nhuận (margin).

**Thông tin công ty (Company Info)**

````python
df = ref.company('TCB').info()
display(df.head() if hasattr(df, 'head') else df)
````

**Danh sách cổ đông lớn (Major Shareholders)**

````python
df = ref.company('VIC').shareholders()
display(df.head() if hasattr(df, 'head') else df)
````

**Ban lãnh đạo / Quản lý cấp cao (Officers)**

````python
df = ref.company('HPG').officers()
display(df.head() if hasattr(df, 'head') else df)
````

**Danh sách công ty con (Subsidiaries)**

````python
df = ref.company('VNM').subsidiaries()
display(df.head() if hasattr(df, 'head') else df)
````

**Tin tức công ty (Company News)**

````python
df = ref.company('TCB').news()
display(df.head() if hasattr(df, 'head') else df)
````

**Lịch sự kiện công ty (Company Events)**

````python
df = ref.company('TCB').events()
display(df.head() if hasattr(df, 'head') else df)
````

### 2. Equity Domain (Danh Sách Cổ Phiếu)

Tra cứu và phân loại danh sách các mã cổ phiếu đang niêm yết trên thị trường.

**Toàn bộ danh sách cổ phiếu (All Equities)**

````python
df = ref.equity.list()
df.head()
````

**Cổ phiếu theo nhóm/rổ chỉ số (Equities by Group)**

````python
df = ref.equity.list_by_group('VN30')
df.head()
````

**Cổ phiếu theo phân từng sàn (Equities by Exchange)**

````python
df = ref.equity.list_by_exchange()
display(df.head() if hasattr(df, 'head') else df)
````

**Cổ phiếu theo phân ngành ICB (Equities by Industry)**

````python
df = ref.equity.list_by_industry()
display(df.head() if hasattr(df, 'head') else df)
````

### 3. Index Domain (Danh Sách Chỉ Số)

Truy xuất danh mục rổ chỉ số, các nhóm chỉ số hỗ trợ và các mã thành phần cấu thành.

**Toàn bộ danh sách rổ chỉ số (All Indices)**

````python
df = ref.index.list()
display(df.head() if hasattr(df, 'head') else df)
````

**Các nhóm rổ chỉ số (List Index Groups)**

````python
df = ref.index.groups()
display(df.head() if hasattr(df, 'head') else df)
````

**Các thành phần trong rổ chỉ số cố định (Index Members)**

````python
df = ref.index.members('VN30')
display(df.head() if hasattr(df, 'head') else df)
````

**Thông tin tổng quan về một rổ chỉ số (Index Info)**

````python
df = ref.index('VN30').info()
display(df.head() if hasattr(df, 'head') else df)
````

### 4. Industry Domain (Ngành Kinh Tế)

Tra cứu các ngành kinh tế chuẩn ICB và phân bổ cổ phiếu trong ngành.

**Danh sách ngành ICB (ICB Industry List)**

````python
df = ref.industry.list()
display(df.head() if hasattr(df, 'head') else df)
````

**Cổ phiếu theo nhóm ngành (Stocks grouped by Sectors)**

````python
df = ref.industry.sectors()
display(df.head() if hasattr(df, 'head') else df)
````

### 5. Fund Domain (Quỹ Đầu Tư Mở)

Khám phá danh sách các Quỹ mở (Mutual Funds) đang hoạt động.

**Danh sách quỹ đầu tư mở (List of Mutual Funds)**

````python
df = ref.fund.list()
display(df.head() if hasattr(df, 'head') else df)
````

### 6. ETF Domain (Quỹ Hoán Đổi Danh Mục)

Truy cập thông tin danh sách các Quỹ Hoán Đổi Danh Mục (ETF) niêm yết.

**Danh sách quỹ ETF (List of ETFs)**

````python
df = ref.etf.list()
display(df.head() if hasattr(df, 'head') else df)
````

### 7. Bond Domain (Trái Phiếu)

Liệt kê danh sách các trái phiếu chính phủ và trái phiếu doanh nghiệp trên thị trường.

**Toàn bộ trái phiếu (All Bonds)**

````python
df_all = ref.bond.list(bond_type='all')
df_all
````

**Trái phiếu Doanh nghiệp (Corporate Bonds)**

````python
df_corp = ref.bond.list(bond_type='corporate')
df_corp
````

**Trái phiếu Chính phủ (Goverment Bonds)**

````python
df_corp = ref.bond.list(bond_type='government')
df_corp
````

### 8. Events Domain (Sự Kiện Thị Trường)

Lịch sự kiện thị trường chung như ngày giao dịch không hưởng quyền, đại hội cổ đông, thay đổi vốn, nghỉ lễ.

**Lịch giao dịch / Sự kiện đặc biệt toàn thị trường (Market-wide Events)**

````python
df = ref.events.market()
df
````

### 10. Futures Domain (Hợp Đồng Tương Lai)

Danh sách và thông tin chi tiết các hợp đồng tương lai chỉ số (Phái sinh).

**Liệt kê danh sách HĐTL (List of Futures Contracts)**

````python
df = ref.futures().list()
df
````

### 11. Warrant Domain (Chứng Quyền Có Bảo Đảm)

Danh mục niêm yết và chi tiết tham số của các Chứng quyền có bảo đảm (Covered Warrants).

**Liệt kê chứng quyền niêm yết (List of Covered Warrants)**

````python
df = ref.warrant().list()
df
````

## 4. Tra cứu Thông tin & Ký hiệu Quốc tế (Global Search)

Sử dụng domain `Search` để tra cứu thông tin meta-data của các tài sản đa quốc gia (Hàng hóa, Ngoại hối, Crypto).

- `search.symbol()`: Quét toàn bộ metadata nội bộ của FXSB nhằm lấy danh sách hàng ngàn instrument hợp lệ với tốc độ mili-giây.
- `search.info()`: Cổng định tuyến mapping với MSN (Microsoft Finance) cung cấp tên đầy đủ, mô tả và sàn giao dịch cho tài sản quốc tế.

````python
print("-- 1. Tìm symbol EUR thông qua MSN Listing --")
df_search_symbol = ref.search.symbol("EUR")
display(df_search_symbol.head())

print("-- 2. Tìm metadata chi tiết mã AAPL thông qua MSN Global Info --")
df_asset_info = ref.search.info("AAPL")
display(df_asset_info.head())
````
