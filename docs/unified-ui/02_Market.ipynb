````python
from vnstock import Market

mkt = Market()
````

## 1. Equity Market / Thị Trường Cổ Phiếu

Retrieve OHLCV, trade history, order book, and other statistics for individual stocks.

*Lấy giá lịch sử, lịch sử lệnh, sổ lệnh, và các thống kê khác cho từng cổ phiếu.*

### Historical Price (OHLCV) / Lịch sử giá

````python
mkt.equity("ACB").ohlcv(length=10, interval='1D')
````

````python
df_ohlcv = mkt.equity("ACB").ohlcv(start="2026-04-01", end="2026-04-10")
df_ohlcv
````

### Intraday Trades / Bảng Chi tiết Lệnh (Time & Sales)

````python
df_trades = mkt.equity("ACB").trades()
df_trades
````

### Quote / Bảng giá

````python
mkt.equity(["ACB", "VCB"]).quote()
````

````python
quote = mkt.equity("HPG").quote()
quote
````

## 2. Index Market / Thị Trường Chỉ Số

Provides historical index points, current quotes and summary. 

*Cung cấp điểm chỉ số lịch sử, bảng giá hiện hành và cấu trúc tổng quan.*\n
\n
> **Note:** Từ phiên bản 3.1.0, Vnstock đã hỗ trợ lấy thông tin thống kê lịch sử giao dịch mở rộng của các loại chỉ số thông qua hàm trade_history().

````python
df_vnindex = mkt.index("VNINDEX").ohlcv(start="2026-01-01", end="2026-03-01")
df_vnindex.head()
````

## 3. Futures & Warrant Market / Thị trường Phái sinh & Chứng quyền

Access futures and warrant markets directly using their respective symbols.

*Truy cập trực tiếp thị trường hợp đồng tương lai và chứng quyền thông qua mã.*

### Futures Market / Hợp Đồng Tương Lai

````python
symbol_future = "VN30F1M"
df_vn30f = mkt.futures(symbol_future).ohlcv(start="2026-04-01", end="2026-04-27")
display(df_vn30f.head())

quote_vn30f = mkt.futures(symbol_future).quote()
quote_vn30f
````

### Warrant Market / Chứng Quyền

````python
symbol_warrant = "CACB2511"

df_warrant = mkt.warrant(symbol_warrant).ohlcv(start="2026-02-01", end="2026-03-01")
display(df_warrant.head())

quote_warrant = mkt.warrant(symbol_warrant).quote()
quote_warrant
````

## 4. ETF & Fund Market / Thị Trường Quỹ

Retrieve data for ETFs and open-ended mutual funds.

*Lấy dữ liệu cho các chứng chỉ quỹ ETF và quỹ đầu tư mở.*

````python
# ETF Historical / Giá lịch sử ETF
df_etf = mkt.etf("E1VFVN30").ohlcv(start="2026-02-01", end="2026-03-01")
display(df_etf.head())

quote_etf = mkt.etf("E1VFVN30").quote()
quote_etf
````

### Mutual Fund Market / Quỹ đầu tư mở

````python
# Mutual Fund NAV / Lịch sử NAV Quỹ mở
df_nav = mkt.fund("VFIBS").history()
display(df_nav.head())

# Top Holding / Tỷ trọng phân bổ tài sản cao nhất
top_holding = mkt.fund("VFIBS").top_holding()
display(top_holding)

# Industry Holding / Quản lý theo nhóm ngành
industry = mkt.fund("VFIBS").industry_holding()
display(industry)
````

## 5. Market Wide / Bảng Giá Nhiều Mã

Efficiently query multiple symbols at once to fetch a real-time snapshot market board.

*Truy vấn nhiều mã cùng lúc một cách tối ưu để lấy tổng quan bảng giá thị trường.*

````python
df_quotes = mkt.quote(["VIC", "TCB", "HPG", "VNM", "BID"])
df_quotes
````

## 3. Crypto

Sử dụng bộ Unified UI để lấy dữ liệu Spot từ thị trường Crypto (nguồn: sàn **Binance**). Hỗ trợ các endpoints như OHLCV, Sổ lệnh (Order Book), Lịch sử giao dịch (Intraday), và Báo giá nhanh (Quote).

````python
from typing import Any
from vnstock import Market
from IPython.display import display

# Khởi tạo đối tượng Market và lấy dữ liệu cặp tiền số BTCUSDT qua domain Crypto
mkt = Market()
crypto = mkt.crypto('BTCUSDT')

print("-- 2. OHLCV Lịch sử (5 Nến ngày gần nhất) --")
crypto.ohlcv(interval='1d', limit=5)
````

## 4. Khám phá thị trường Ngoại hối & Hàng hoá Quốc tế

Các tham số:
- `interval`: Hỗ trợ chuẩn `1m`, `1h`, `4h`, `1d`, `1w`, `1M`. Có cơ chế gom nến tự động.
- `side`: Tùy chọn lấy giá `bid` (mặc định), `ask` hoặc `mid`.


````python
print("-- 1. Tỷ giá Ngoại Hối EUR/USD (Interval 1 giờ, 5 nến) --")
df_forex = mkt.forex("EURUSD").ohlcv(interval="1h", length=5, side="mid")
display(df_forex.head())
````

````python
print("-- 2. Lịch sử giá Vàng XAU/USD (Interval 1d, 5 nến) --")
df_gold = mkt.forex("XAUUSD").ohlcv(interval="1d", length=5)
display(df_gold)
````

````python
print("-- 3. Chỉ số Chứng khoán Dow Jones US30 (Interval 4h, 5 nến) --")
# Hỗ trợ mã ngắn USA30, thư viện sẽ tự động phân giải thành USA30.IDX-USD
df_dowjones = mkt.index("USA30", scope="global").ohlcv(interval="4h", length=5)
display(df_dowjones)

print("-- 4. Đổi Múi Giờ Dữ Liệu Thị Trường Quốc Tế --")
# Dukascopy mặc định trả dữ liệu tính theo chuẩn khung giờ Việt Nam (GMT+7).
# Bạn có thể sử dụng param timezone để đưa dữ liệu về trả chuẩn UTC (hoặc bất kỳ múi giờ nào khác)
df_utc = mkt.forex("EURUSD", timezone="UTC").ohlcv(interval="1d", length=5)
display(df_utc)
````
